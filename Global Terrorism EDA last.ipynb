**4. Attack types and there success rates in India** 

```python
succesrates = dfind.groupby(by='attacktype1_txt')['success'].value_counts()
suc = pd.DataFrame(columns=['Attacktype','Successeded','Failed'],index=[0,1,2,3,4,5,6,7,8])
```

```python
atacktp = ['Hijacking', 'Bombing/Explosion', 'Armed Assault', 'Assassination',
       'Facility/Infrastructure Attack', 'Unknown',
       'Hostage Taking (Kidnapping)', 'Unarmed Assault',
       'Hostage Taking (Barricade Incident)']
succeded = [] 
non_succeded = []
```

```python
for i in atacktp:
  ## For Succeded attacks 
  fd = dfind[(dfind['attacktype1_txt']==i) & (dfind['success']==1)]
  cnt = len(fd['success'])
  succeded.append(cnt)

  ## For Non Succeded attacks
  fd = dfind[(dfind['attacktype1_txt']==i) & (dfind['success']==0)]
  cnt = len(fd['success'])
  non_succeded.append(cnt)
```

```python
suc['Attacktype'] = atacktp
suc['Successeded'] =succeded
suc['Failed'] = non_succeded
suc['Total'] = suc['Successeded'] + suc['Failed']
suc['Success_Per'] = round((suc['Successeded'] / (suc['Successeded'] + suc['Failed'])) *100,2)
```

```python
suc.reset_index(inplace=True)
```

```python
suc.drop('index',axis=1,inplace=True) 
```

```python
plt.rcParams['figure.figsize'] = [14,8]
plt.rcParams['figure.autolayout'] = True

fig, ax1 = plt.subplots()

plots = sns.barplot(x='Attacktype', y='Total', data=suc)
plots.set_xticklabels(plots.get_xticklabels(), rotation=40, ha="right")
 
# Iterating over the bars one-by-one
for bar in plots.patches:
    plots.annotate(format(bar.get_height(), '.0f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=13, xytext=(0, 8),
                   textcoords='offset points')
 
# Setting the label for x-axis
plt.xlabel("Attack Type", size=14)
 
# Setting the label for y-axis
plt.ylabel("Total Terror Attacks", size=14)
 
# Setting the title for the graph
plt.title("Total Attacks With Respect To Attack Type In India",size=16)

sns.lineplot(data =suc['Successeded'], marker='*',color='black',sort = True,label='Successful Attacks')

# Finally showing the plot
plt.show()
```

1) Data is nominal categorical and we want to display counts of attacks per attack type so bar chart is suitable for such analysis.

2) Chart shows bombing and explosion, Armed assualt, assassination and hostage taking are few of the attack types that happened in India.

```python
suc.Success_Per.mean()
```

I) Total 11960 terrorist attacks happened in India from 1972 to 2017, Out of which 92% attacks were successful.

II) Hostage Taking (Kidnapping),Unarmed Assault this type of attack types have highest success rate around 99%.
